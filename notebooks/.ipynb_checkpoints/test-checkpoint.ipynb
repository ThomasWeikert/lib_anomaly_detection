The approach described here is to detect outliers in hotel occupancy data by estimating a Poisson rate parameter for the expected number of arrivals and using the lognormal distribution to estimate the confidence interval around this parameter. To do this, the data is preprocessed and grouped by arrival date, generating a daily arrivals dataframe. Sequences of 30 consecutive days of arrivals are created, and the expected arrival rate for each sequence is estimated as the average number of arrivals per day. The standard deviation of the arrival rate for each sequence is also calculated. The lognormal distribution is then used to estimate the 98% confidence interval for the expected arrival rate for each sequence. Finally, a function is defined to identify outliers as any observation whose arrival rate is below the 98% confidence interval of the expected arrival rate.

```python
import pandas as pd
import numpy as np
from scipy.stats import lognorm
from datetime import datetime, timedelta
import matplotlib.pyplot as plt
import warnings

# Set the warning filter to "ignore"
warnings.filterwarnings("ignore")


def load_data(filepath):
    """Loads data from a CSV file into a Pandas dataframe."""
    data = pd.read_csv(filepath)
    return data

def preprocess_data(data):
    """Preprocesses the input data, generating a subset of columns and new features."""
    subset = data[['hotel','arrival_date_year','arrival_date_month','arrival_date_day_of_month','lead_time','adults','children']]
    subset.loc[:, 'pax'] = [adults + children for adults, children in zip(subset['adults'],subset['children'])]
    subset.loc[:, 'arrival_dt'] = [datetime(int(year), datetime.strptime(month, "%B").month, int(day)) for year, month, day in zip(subset['arrival_date_year'], subset['arrival_date_month'], subset['arrival_date_day_of_month'])]
    subset.loc[:, 'booking_dt'] = [dt - timedelta(days = lead_time) for dt, lead_time in zip(subset['arrival_dt'], subset['lead_time'])]
    subset = subset.drop(columns = ['arrival_date_year','arrival_date_month','arrival_date_day_of_month','lead_time','adults','children'])
    return subset

def create_continuous_and_daily_grouped_df(subset):
    """Groups the preprocessed data by arrival date, generates a continuous date sequence, and fills missing values with zeroes."""
    subset = subset.drop(columns = ['hotel'])
    data_grouped = subset.groupby('arrival_dt')[['pax']].sum()
    date_range = pd.date_range(start=subset['arrival_dt'].min(), end=subset['arrival_dt'].max(), freq='D')
    continuous_df = pd.DataFrame({'arrival_dt': date_range})
    merged_df = pd.merge(data_grouped, continuous_df, on='arrival_dt', how='right')
    merged_df = merged_df.reindex(columns = ['arrival_dt', 'pax'])
    merged_df = merged_df.rename(columns = {'arrival_dt':'date'})
    merged_df['pax'] = merged_df['pax'].fillna(0)
    return merged_df

def get_hotel_arrivals_by_date(data, hotel_name):
    """Filters the preprocessed data by hotel name, generates a daily arrivals dataframe, and returns it."""
    hotel_data = data[data['hotel'] == hotel_name]
    daily_arrivals = create_continuous_and_daily_grouped_df(hotel_data)
    return daily_arrivals

def create_sequences(df):
    sequences = []
    for i in range(14, len(df), 1):
        sequence = df.iloc[i-14:i]['pax']
        sequences.append(sequence)
    return pd.array(sequences)

def get_next_values(df):
    next_values, next_values_date  = [], []
    for i in range(14, len(df), 1):
        next_value = df.iloc[i]['pax']
        next_value_date = df.iloc[i]['date']
        next_values.append(next_value)
        next_values_date.append(next_value_date)
    return pd.array(next_values), pd.array(next_values_date)

def calculate_exp_rate(seq_arr):
    exp_rates = []
    for i in range(len(seq_arr)):
        exp_rate = np.floor(sum(seq_arr[i]) / len(seq_arr[i]))
        exp_rates.append(exp_rate)
    return exp_rates

def calculate_exp_rate_std(seq_arr):
    exp_rate_stds = []
    for i in range(len(seq_arr)):
        exp_rate_std = np.std(seq_arr[i])
        exp_rate_stds.append(exp_rate_std)
    return exp_rate_stds

def make_lognorm(mean,std):
    # Creates lognormal distribution with specific mean and standard deviation
    shape = np.sqrt(np.log((std**2 + mean**2)/mean**2))
    scale = (mean**3 * np.sqrt((std**2 + mean**2)/mean**2))/(std**2 + mean**2)
    return stats.lognorm(shape,scale=scale)

def get_exp_rate_intervals(exp_rates, exp_rate_stds):
    exp_rate_ci_mins, exp_rate_ci_maxs = [], []
    for i in range(len(exp_rates)):
        exp_rate_ci_min, exp_rate_ci_max = make_lognorm(exp_rates[i], exp_rate_stds[i]).interval(0.98)
        exp_rate_ci_mins.append(exp_rate_ci_min)
        exp_rate_ci_maxs.append(exp_rate_ci_max)
    return exp_rate_ci_mins, exp_rate_ci_maxs

def is_outlier(exp_rate_ci_min, obs):
    obs_rate = est_poission_rate_ci(obs,1)
    return obs_rate.cdf(exp_rate_ci_min) > 0.99
    
def est_poission_rate_ci(n_events, 
                         source_duration, 
                         target_duration=1.0, 
                         ci=0.98,
                         expected_avg_n_events_per_unit_duration=200):
    """
    Estimate poission rate parameter confidence interval

    Parameters
    ----------
    n_events : int
        number of observed events
    source_duration : float
        duration for which n_events were observed
    target_duration : float
        duration for which to return expected rate
    ci : float
        size of confidence interval

    Returns
    -------
    (float,float)
        Confidence interval left, right bounds
    """
    
    def calc_posterior(prior_shape, prior_scale, n_events, duration):
        posterior_shape = prior_shape + n_events
        posterior_scale = prior_scale/(prior_scale*duration+1)

        return posterior_shape, posterior_scale

    rate_prior_average = expected_avg_n_events_per_unit_duration
    rate_prior_shape = 1
    rate_prior_scale = rate_prior_average/rate_prior_shape
    rate_prior = stats.gamma(rate_prior_shape,scale=rate_prior_scale)
    
    posterior_shape, posterior_scale = calc_posterior(rate_prior_shape, 
                                                      rate_prior_scale, 
                                                      n_events, 
                                                      source_duration/target_duration)
    posterior = stats.gamma(posterior_shape,scale=posterior_scale)
    return posterior

def plot_outliers(results):
    outliers = results[results['is_outlier']]
    plt.title("Number of events vs confidence interval")
    plt.ylabel("Number of events")
    _ = results
    _['n_obs_events'].plot()
    plt.fill_between(_.index, 
                     _['ci_min'], 
                     _['ci_max'],
                     color='k',
                     lw=0,
                     alpha=0.2)
    plt.scatter(outliers.index, outliers['n_obs_events'], marker='x', color='r')
    plt.show()

def create_sequences_and_outliers(df):
    sequences = create_sequences(df)
    next_values, next_values_date = get_next_values(df)
    exp_rates = calculate_exp_rate(sequences)
    exp_rate_stds = calculate_exp_rate_std(sequences)
    exp_rate_ci_mins, exp_rate_ci_maxs = get_exp_rate_intervals(exp_rates, exp_rate_stds)
    is_outliers = [is_outlier(exp_rate_ci_min, obs) for exp_rate_ci_min, obs in zip(exp_rate_ci_mins, next_values)]
    return sequences, next_values, next_values_date, exp_rates, exp_rate_ci_mins, exp_rate_ci_maxs, is_outliers


data = load_data("hotel_booking.csv")
subset = preprocess_data(data)
resort_hotel_arrivals = get_hotel_arrivals_by_date(subset, "Resort Hotel")
city_hotel_arrivals = get_hotel_arrivals_by_date(subset, "City Hotel")

# city hotel
sequences, next_values, next_values_date, exp_rates, exp_rate_ci_mins, exp_rate_ci_maxs, is_outliers = create_sequences_and_outliers(city_hotel_arrivals)
data = {'date': next_values_date, 'n_obs_events': next_values, 'exp_rate': exp_rates, 'ci_min': exp_rate_ci_mins, 'ci_max': exp_rate_ci_maxs, 'is_outlier': is_outliers}
results_city = pd.DataFrame(data)
print(results_city.head())
plot_outliers(results_city)


# resort hotel
sequences, next_values, next_values_date, exp_rates, exp_rate_ci_mins, exp_rate_ci_maxs, is_outliers = create_sequences_and_outliers(resort_hotel_arrivals)
data = {'date': next_values_date, 'n_obs_events': next_values, 'exp_rate': exp_rates, 'ci_min': exp_rate_ci_mins, 'ci_max': exp_rate_ci_maxs, 'is_outlier': is_outliers}
results_resort = pd.DataFrame(data)
print(results_resort.head())
plot_outliers(results_resort)
```
